```python
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
diamonds = pd.read_csv('diamonds.csv',index_col=0)
```

### Task 1: Find all the diamonds with more than 2 carat and plot their price distribution in a histogram

```python
carat = diamonds[diamonds['carat']>2]
carat["price"].plot(kind="hist")
```

### Task 2: compute and plot the standard deviation of the x dimension for the different cuts

```python
ca = diamonds.groupby("cut").std()
ca[['x']].plot()
```

### Task 5: Plot the age composition of the different Bundesländer

```python
bund.groupby("NAME")["ALTER_1","ALTER_2","ALTER_3","ALTER_4","ALTER_5"].sum().plot(kind="barh", stacked=True)
```

### Task 6: ...and now relative to the population in each Bundesland

```python
rel = bund[['ALTER_1','ALTER_2','ALTER_3','ALTER_4','ALTER_5']].div(bund['EWZ'], axis=0)
print rel
#print rel.index

#print rel.index
rel.groupby(bund['NAME'])["ALTER_1","ALTER_2","ALTER_3","ALTER_4","ALTER_5"].sum().plot(kind="barh", stacked=True)
```
